## 廃止されたプロセッサーファミリー
- **Canary**: Canaryファミリーは、5量子ビットから16量子ビットまでの小規模な設計で構成されていました。
- **Falcon**: Falconファミリーのデバイスは、中規模回路向けの優れたプラットフォームを提供し、より大型のデバイスへ展開する前にパフォーマンスとスケーラビリティの改善を実証する場としても活用されました。
- **Egret**: Egretは、チューナブルカプラーの革新技術を33量子ビットのプラットフォームに導入し、より高速かつ高忠実度の2量子ビットゲートを実現しました。
- **Hummingbird**: ヘビー・ヘキサゴナル量子ビットレイアウトを採用したHummingbirdファミリーは、最大65量子ビットをサポートしました。
- **Osprey**: OspreyはEagleのほぼ4倍の規模となる433量子ビットを備えていました。チップサイズの大型化に伴い、デバイスパッケージングのさらなる改善や、同一配線フットプリント内でより多くのI/O要件に対応するためのクライオスタット内カスタムフレキシブルケーブルが必要でした。
- **Eagle**: Eagleプロセッサーファミリーは、それ以前の世代よりもスケーラブルなパッケージング技術を採用しました。特に、信号が複数のチップ層を通過することで、性能を損なうことなく高密度のI/Oを実現しました。
